```python
import json
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

### Top frequent tags

```python
data = json.load(open("../data/questions_100k_full.json"))

tag_count = {}
for question in data["items"]:
    for tag in question["tags"]:
        if tag not in tag_count:
            tag_count[tag] = 1
        else:
            tag_count[tag] += 1

n_questions = len(data["items"])
print(n_questions, len(tag_count))
```

```python
tags = []
props = []
for k, v in sorted(tag_count.items(), key=lambda item: item[1], reverse=True):
    tags += [k]
    props += [v / n_questions]
```

```python
N = 20

# https://matplotlib.org/3.1.1/gallery/lines_bars_and_markers/barh.html
plt.rcdefaults()
fig, ax = plt.subplots()
y_pos = np.arange(N)
ax.barh(y_pos, props[:N], align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(tags[:N])
ax.invert_yaxis()
ax.set_xlabel('Proportion of questions')
plt.show()
```

### Distribution of years

```python
year_count = {}
for question in data["items"]:
    year = datetime.fromtimestamp(question["creation_date"]).year
    if year not in year_count:
        year_count[year] = 1
    else:
        year_count[year] += 1
```

```python
years = []
props = []
for k, v in sorted(year_count.items(), key=lambda item: item[1], reverse=True):
    years += [k]
    props += [v / n_questions]
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()
y_pos = np.arange(len(years))
ax.barh(y_pos, props, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(years)
ax.invert_yaxis()
ax.set_xlabel('Proportion of questions')
plt.show()
```
