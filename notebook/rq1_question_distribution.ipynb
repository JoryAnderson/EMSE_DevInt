RQ1: What are the distributions of developers that post questions?

```python
import matplotlib.pyplot as plt

%load_ext sql
%sql sqlite:///uname_qid.db

query = """ WITH uname_c AS (
                SELECT uname, COUNT(*) AS c
                FROM uname_qid
                GROUP BY uname)
            SELECT c, COUNT(*)
                FROM uname_c
                GROUP BY c"""

res = %sql $query
df = res.DataFrame()
df.columns = ["# questions", "# developers"]
```

```python
df
```

Plot from # questions 1-37, as the numbers afterwards are discrete.

```python
plt.xlim(1, 37)
plt.yscale('log', basey=10)
plt.xlabel("Number of posted questions")
plt.ylabel("Number of developers")
plt.plot(df["# questions"], df["# developers"])
plt.show()
```
